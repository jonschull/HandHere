```
#DEMO #https://python-forum.io/Thread-Pasting-multiple-images-on-to-one-image-with-a-for-loop
from PIL import Image
 
img_to_paste = Image.open('msg.png')
width, height = img_to_paste.size
 
n = 2   # paste image n-times
 
img = Image.new('RGB', (width, height*n), color=(0, 0, 0))     # set width, height of new image based on original image
img.save('out.png')
 
for i in range(0, n):
    out = Image.open('out.png')
    out.paste(img_to_paste, (0,i*width))    # the second argument here is tuple representing upper left corner
    out.save('out.png')
    
from IPython.display import Image
Image('out.png')
```

```
Prefix="Re-NABLE"

def makeName(number):
  return Prefix + '_' + str(number)

name = makeName(13)
pngName = name+'.png'

def makeMessage(name):
  msg=f"""MATMSG:TO:jschull@gmail.com;SUB:Send us a selfie with device {name}!;BODY:Please send us a selfie!   
  We like to know where these devices go and how they are used!
  In the future, you'll be able to track this device at http://wikifactory.com/+e-NABLE/projects/HandHere.
"""
  return msg

msg = makeMessage(name)
print(msg)

#!pip3 install pypng
#!pip3 install pyqrcode; 

import pyqrcode
import PIL
import IPython.display
from PIL import ImageDraw
from PIL import ImageFont

def getQRpng(msg):
  QRcode = pyqrcode.create(msg)
  QRcode.png(pngName,scale=3)
  return PIL.Image.open(pngName)

QRpng = getQRpng(msg)

def addCaption(QRpng=QRpng, name=name, pngName=pngName):
  width, height = QRpng.size
  caption = PIL.Image.new('RGB', (width, height + 20), color=(255, 255, 255))

  fnt = ImageFont.truetype("arial.ttf", 15)
  d = ImageDraw.Draw(caption)
  d.text((width/4,height), name,  font=fnt,  fill=(0,0,0))


  caption.paste(QRpng,(0,0))
  caption.save(pngName)

addCaption(QRpng, name, pngName)
captionPng = PIL.Image.open(pngName)
IPython.display.Image(pngName)
```
